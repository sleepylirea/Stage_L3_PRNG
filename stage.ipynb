```python
def racine2(N): # calcule N chiffres de sqrt(2)
  # pour cela on encadre : k^2 <= 2 10^2N < (k+1)^2
  d = 2 * 100**N
  k = int(sqrt(2)*10**N)
  while k**2 <= d:
    k += 1
  return k-1

racine2(10**2)
```

```python
def isPrime(n) :
    for i in range(2,(n//2)+1) :
        if (n%i) == 0 :
            return False
    return True

def max(L,p):
    x = L[0]
    for i in range (p):
        if L[i]>x :
            x = L[i]
    return x
```

```python
def power(a,n,d): # a^n mod d
    if n == 0:
        return 1
    r = power(a,n//2,d)
    r = (r*r) % d
    if n % 2 == 0: # n pair
        return r
    # n impair
    return (r*a) % d

def invMod(a,p): # inverse de a modulo p
    return power(a,p-2,p)


#for k in range(100000):
#    invMod(3,10001)

```

```python
def craquerWMod(L,m):
    # nombre = reste de (ax+b) par m
    # e1 - e2 = ax+b -(ae1+b) = ax - ae1 = a(x-e1)
    # e1 - e2 = ax+b -(a(ax+b)+b) = ax - a(ax+b) = -a**2x + ax - ab = e1 - e2
    # => -ax + x - b = (e1-e2)/a   =>   (a-1)x + b = -(e1-e2)/a

    x = L[0]
    e1 = L[1]         # e1 = a * L[0] + b
    e2 = L[2]         # e2 = a * L[1] + b

    # e1 - e2 = a * L[0] + b - (a * L[1] + b)
    # a = (e1-e2)/(e1-x)
    a = ((e1-e2)*invMod((x-e1),m))%m

    b = (e1 - x*a)%m

    # print ("x = "+str(x)+"\na = "+str(a)+"\nb = "+str(b))
    # print (f"x = {x}\na = {a}\nb = {b}")
    return (x,a,b)
    
```

```python
def uplet5 (L,a) : 
    for i in range (a):
        n = len(L) - 5
        L.append((L[n+2]+L[n])%2)
    return L

uplet5 ([0,0,0,0,1],20)
```

```python
def LecB5(L) : 
    B = []
    n = len(L)
    for i in range (n-5):
        c = []
        for j in range (5):
            c.append(str(L[i+j]))
        a = ''.join(c)
        b = int(str(a),2)
        B.append(b)
    return B
```
