## Prévision pierre feuille ciseau

```python
from random import *

def PFC ():
    x = random()
    while x == 0 :
        x = random()
    return "pierre" if x <= 1/3 else "feuille" if x <= 2/3 else "ciseaux"

PFC()
```

```python
def compter(L):
    D = {}
    for i in range (len(L)):
        if L[i] not in D :
            D[L[i]] = 1
        else : 
            D[L[i]] += 1
    return D

def compterP(L):
    D = compter(L)
    for k in D.keys():
        D[k] = D[k]/len(L)
    return D
```

```python
L = [PFC() for x in range (300000)]
#print(L)
compterP(L)
```

```python
def PFCprev (L) :    # list => [(v,p),(v,p),(d,p),(d,f),...]
    Ld = []
    for i in range (len(L)):
        Ld.append(L[i][1])
    Dp = compterP(Ld)
    return [Dp["c"],Dp["c"]+Dp["p"]]       # renvoie pourcentage pour prevision de ce qu'on doit jouer
```

```python
def PFCprevu (Lp):
    x = random()
    while x == 0 :
        x = random()
    return "pierre" if x <= Lp[0] else "feuille" if x <= Lp[1] else "ciseaux"

Lp = [1/5,2/5]
compterP([PFCprevu(Lp) for x in range (300000)])
```

```python
def PFCprobaVD (L):            # tuple (v,p) => (victoire, pierre)
    D = {"v" : [0,0], "d" : [0,0]}
    Nb = [0,0]
    for i in range (1,len(L)):
        if L[i-1][0] == "v" :
            Nb[0] += 1
            if L[i][1] == L[i-1][1] :
                D["v"][0] += 1
            else :
                D["v"][1] += 1
        elif L[i-1][0] == "d" :
            Nb[1] += 1
            if L[i][1] == L[i-1][1] :
                D["d"][0] += 1
            else :
                D["d"][1] += 1
    Lproba = [D["v"],D["d"]]
    for i in range (2) :
        Lproba[0][i] = Lproba[0][i]/Nb[0]
    for j in range (2):
        Lproba[1][j] = Lproba[1][j]/Nb[1]

    return Lproba  # [[victoire],[defaite]]  => proba rejouer apres victoire/defaite, ou non

L = [(v,p),(d,p),(d,c),(v,f),(v,f),(v,c),(d,c),(v,c),(v,p),(d,p),(v,c),(d,c),(v,f)]

Lproba = PFCprobaVD (L)
PFCprobaVD (L)
```

```python
def PFCVD (Lproba,tuple):
    t = tuple[1]
    if t == "p" or t == "f" or t == "c" :
        if t == "p":
            t = "pierre"
        elif t == "f":
            t = "feuille"
        else :
            t = "ciseaux"

    x = random()
    
    if tuple[0] == "v" :
        if x < Lproba[0][0]:
            return t
        r = PFC()
        while r == t:
            r = PFC()
        return r
    elif tuple[0] == "d" :
        if x < Lproba[1][0]:
            return t
        r = PFC()
        while r == t:
            r = PFC()
        return r
    
    return PFC()


PFCVD (Lproba,(e,c))
```

```python
L = [PFCVD(Lproba,(d,c)) for x in range (300000)]
compterP(L)
```

```python
L = [PFCVD(Lproba,(v,c)) for x in range (300000)]
compterP(L)
```
